```python
import itertools

def generate_expressions(numbers):
    def helper(operators, current_expression, remaining_numbers, all_expressions):
        if not remaining_numbers:
            all_expressions.append(current_expression)
            return
        
        current_number = remaining_numbers[0]
        remaining_numbers = remaining_numbers[1:]
        
        for operator in operators:
            new_expression = current_expression + operator + str(current_number)
            helper(operators, new_expression, remaining_numbers, all_expressions)
    
    operators = ['+', '-', '*', '/']
    all_expressions = []
    
    for permutation in itertools.permutations(numbers):
        first_number = permutation[0]
        remaining_numbers = permutation[1:]
        
        helper(operators, str(first_number), remaining_numbers, all_expressions)
    
    return all_expressions

# Example usage
numbers = [2, 3, 5]
expressions = generate_expressions(numbers)
print("Generated expressions:")
for expression in expressions:
    print(expression)
```
